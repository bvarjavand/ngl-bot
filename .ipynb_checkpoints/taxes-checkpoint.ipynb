```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fed = {
    612351 : .37,
    408201 : .35,
    321451 : .32,
    168451 : .24,
    78951 : .22,
    19401 : .12
}

ca = {
    1000001 : .133,
    590742 : .123,
    354445 : .113,
    295373 : .103,
    57824 : .093,
    45753 : .08
}
```

```python
def find_taxes(x):
    curr_tax_fed = 0
    curr_tax_ca = 0
    for value, tax in list(fed.items())[::-1]:
        if x > value:
            curr_tax_fed = tax
    for value, tax in list(ca.items())[::-1]:
        if x > value:
            curr_tax_ca = tax    
    tot_tax = curr_tax_fed + curr_tax_ca
    return tot_tax
```

```python
def compute_donation_value(amount, step=100):
    income = np.linspace(1,amount, int(amount/step)).astype(int)
    net_income = [(1-find_taxes(i))*i for i in income]
    max_donation = int(amount/2)
    donation = np.linspace(1, max_donation, int(max_donation/step)).astype(int)
    donation_value = [(net_income[-1]-net_income[-i])/i for i in range(len(donation))]
    return income, net_income, donation, donation_value
```

```python
_, _, _,donation_value = compute_donation_value(800000, 1000)
print(len(donation_value))
```

```python
def generate_donation_array(amount, step=10000):
    arr = np.zeros((int(amount/step), int(amount/step)))
    print(arr.shape)
    j = 0
    for i in np.linspace(1, int(amount/step), int(amount/step)).astype(int):
        _, _, _,donation_value = compute_donation_value(i, step)
        arr[j, :len(donation_value)-1] = donation_value
        j += 1
    return arr
```

```python
arrr = generate_donation_array(800000)
```
